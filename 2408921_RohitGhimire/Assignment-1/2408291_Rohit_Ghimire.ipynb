Task - 3 - Outlier Detection:

Tasks:

Identify outlier countries in South Asia based on their Score and GDP per Capita.

Define outliers using the 1.5 × IQR rule.

Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color.

Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

gdp_outliers = detect_outliers(SouthAsia_df, 'Log GDP per capita')
score_outliers = detect_outliers(SouthAsia_df, 'score')
print("Outliers in GDP per Capita\n",gdp_outliers['Country name'])
print("Outliers in Score:\n",score_outliers['Country name'])
```

```python
plt.scatter(SouthAsia_df['Log GDP per capita'], SouthAsia_df['score'], color='blue',label='Normal countries')
plt.scatter(gdp_outliers['Log GDP per capita'], gdp_outliers['score'], color='red', label='Outliers')
plt.title("Outlier Detection: GDP per Capita vs Score", fontsize=16)
plt.xlabel("GDP per Capita", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.legend()
plt.tight_layout()

plt.show()
```

Task - 4 - Exploring Trends Across Metrics: Tasks:


1. Choose two metrics (e.g., Freedom to Make Life Choices and Generosity) and calculate their correlation {pearson correlation} with the Score for South Asian countries.
2. Create scatter plots with trendlines for these metrics against the Score.
3. Identify and discuss the strongest and weakest relationships between these metrics and the Score for South Asian countries.

```python
import numpy as np
def pearson_correlation(x, y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    numerator = np.sum((x - x_mean) * (y - y_mean))
    denominator = np.sqrt(np.sum((x - x_mean) ** 2) * np.sum((y - y_mean) ** 2))

    return numerator / denominator

freedom_correlation= pearson_correlation(SouthAsia_df['Freedom to make life choices'], SouthAsia_df['score'])
generosity_correlation = pearson_correlation(SouthAsia_df['Generosity'], SouthAsia_df['score'])
print("correlation of Freedom to make life choices to Score is: ", freedom_correlation)
print("correlation of Generosity to Score is: ", generosity_correlation)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(SouthAsia_df['Freedom to make life choices'], SouthAsia_df['score'], color='blue', label='Freedom to make life choices')
plt.title("Scatter Plot of Freedom to make life choices vs Score")
plt.xlabel("Freedom to make life choices")
plt.ylabel("Score")
plt.legend()

plt.show()
```
