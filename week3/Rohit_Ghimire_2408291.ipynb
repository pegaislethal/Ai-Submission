```python
fare_per_year = dataframe["Fare"] / dataframe["Age"]
dataframe["fare_per_year"] = fare_per_year

high_fare_age = dataframe[dataframe["fare_per_year"] > 5]

high_fare_age_srt = high_fare_age.sort_values(by="fare_per_year", ascending=False)

result_df = high_fare_age_srt[["Name", "fare_per_year"]]

print(result_df.head())
```

```python
fare_per_class = dataframe["Fare"] / dataframe["Pclass"]
dataframe["fare_per_class"] = fare_per_class

adult_males = dataframe[(dataframe["Age"] >= 18) & (dataframe["Sex"] == "male")]

adult_males_srt = adult_males.sort_values(by="fare_per_class", ascending=False)

result_df = adult_males_srt[["Name", "Age", "fare_per_class"]]

print(result_df.head())
```

b)

```python
total_fare = dataframe["Fare"].sum()

first_class_fare = dataframe[dataframe["Pclass"] == 1]["Fare"].sum()
second_class_fare = dataframe[dataframe["Pclass"] == 2]["Fare"].sum()
third_class_fare = dataframe[dataframe["Pclass"] == 3]["Fare"].sum()

combined_fare = [first_class_fare, second_class_fare, third_class_fare]

proportions = [fare / total_fare *100 for fare in combined_fare]

print(proportions)
```

```python
import pandas as pd
import numpy as np


conditions = [
    dataframe['Age'] < 18,
    (dataframe['Age'] >= 18) & (dataframe['Age'] < 65),
    dataframe['Age'] >= 65
]

choices = ['child', 'adult', 'senior']

dataframe['age_group'] = np.select(conditions, choices, default='unknown')

print(dataframe[['Age', 'age_group']].head())
```

```python
total_passengers = len(dataframe)

total_passengers
```

```python
age_group_counts = dataframe['age_group'].value_counts()

age_group_counts
```

```python
age_group_proportions = age_group_counts / total_passengers

age_group_proportions
```

```python
age_group_percentage = age_group_proportions * 100

print(f"{age_group_percentage}")
```
